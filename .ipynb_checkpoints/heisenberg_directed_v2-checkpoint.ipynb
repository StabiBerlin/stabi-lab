```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt

def create_directed_network_of_letters(filename):
    df = pd.read_csv(filename, sep=';')
    date = df['date'].values.tolist()
    senders = df['senders'].values.tolist()
    receivers = df['receivers'].values.tolist()
    
    G = nx.DiGraph()
    
    for i in range(len(senders)):
        G.add_node(senders[i], name=senders[i])
        G.add_node(receivers[i])
        G.add_edge(senders[i], receivers[i], date=date[i])  # add date as edge attribute
    
    return G

# Load your graph
G = create_directed_network_of_letters('heisenberg_namesdates.csv')

# Draw it
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G, seed=42)

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_color='skyblue', node_size=800)

# Draw edges with proper arrow display
nx.draw_networkx_edges(
    G,
    pos,
    edge_color='gray',
    arrows=True,
    arrowstyle='->',
    arrowsize=20,
    connectionstyle='arc3,rad=0.1'  # curved edges to make direction clearer
)

# Draw node labels
nx.draw_networkx_labels(G, pos, font_size=10, font_family='sans-serif')

# Extract edge labels (dates)
edge_labels = nx.get_edge_attributes(G, 'date')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8, label_pos=0.5)

plt.title("Directed Network of Letters with Dates")
plt.axis('off')
plt.tight_layout()
plt.show()
```
