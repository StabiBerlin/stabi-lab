```python
## networkx chatgpt

import networkx as nx
import pandas as pd

def create_directed_network_of_letters(filename):
    # loading the data
    df = pd.read_csv(filename, sep=';')
    date = df['date'].values.tolist()
    senders = df['senders'].values.tolist()
    receivers = df['receivers'].values.tolist()
    
    # use DiGraph for directed graph
    G = nx.DiGraph()
    
    # add nodes and directed edges
    for i in range(len(senders)):
        G.add_node(senders[i], date=date[i], name=senders[i])
        G.add_node(receivers[i])
        G.add_edge(senders[i], receivers[i], date=date[i])  # optional: include date as edge attribute
    
    return G

# Create the directed graph
G = create_directed_network_of_letters('heisenberg_namesdates.csv')

```

```python
import matplotlib.pyplot as plt

# draw the network
plt.figure(figsize=(12, 12))
#pos = nx.spring_layout(G, seed=42)
pos = nx.fruchterman_reingold_layout(G, seed=42)


# draw nodes
nx.draw_networkx_nodes(G, pos, node_color='skyblue', node_size=700)

# draw directed edges with arrows
nx.draw_networkx_edges(G, pos, edge_color='gray', arrows=True, arrowstyle='->')

# draw labels
nx.draw_networkx_labels(G, pos, font_size=10, font_family='sans-serif')

# show plot
plt.title("Directed Network of Letters")
plt.axis('off')
plt.show()
```
